```python
def strategy(data):
    minimum=data['Open'].iloc[0]
    maximum=data['Open'].iloc[0]
    signals=[]
    signals.append(0)
    for i in range(1,len(data)):
        if(data['Open'].iloc[i]<=minimum):
            signals.append(-1)
            minimum=data['Open'].iloc[i]
        elif(data['Open'].iloc[i]>=maximum):
            signals.append(1)
            maximum=data['Open'].iloc[i]
        else:
            signals.append(0)
    data['signals']=signals
    return data
```

```python
apple=strategy(apple)
apple
```
